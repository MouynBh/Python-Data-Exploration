# **Exploratory Data Analysis Project: Inflation Analysis (1970-2022)**

Welcome to this exploratory data analysis project where we'll dive into understanding a dataset through careful observation. We'll begin by gathering essential information like what's in the data, by looking at its start and end (head and tail), and also by counting how many different things there are (unique values) and what might be missing (missing values).

To make our data more complete, we'll replace the missing information with the average of other known numbers. Additionally, we'll clean up the dataset by removing columns that don't give us valuable insights.

Next, we'll move on to visualization. We'll create graphs and charts to help us see how the prices have changed over time (inflation) for each year and for each part of the world (continent).

Lastly, we'll build a special table that shows us the countries with the highest and lowest inflation rates in the years 1970 and 2022. This will give us a clear picture of how inflation has affected different countries over time. Let's get started on this exciting journey of exploring and understanding our data!

**Importing Essential Python Libraries for Data Analysis and Visualization :**

```python
# Import packages:
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import re
```

```python
data = pd.read_csv('/content/Global Dataset of Inflation.csv', encoding='iso-8859-1')
```

```python
# We check our data infos
data.info()
```

```python
# get a list of all the columns in the dataset
columns_list = data.columns.tolist()

print(columns_list)
```

```python
# Convert column names to lowercase
data.columns = [col.lower() for col in data.columns]
# Replace spaces with underscores (_)
data.columns = [col.replace(' ', '_') for col in data.columns]
# Replace special characters with underscores (_)
data.columns = [re.sub(r'\W+', '_', col) for col in data.columns]
```

```python
# Define the columns for which you want to find unique values
columns_to_check = ['indicator_type', 'note', 'country_code', 'series_name']

# Loop through the columns and print unique values
for column in columns_to_check:
    unique_values = list(data[column].unique())
    print(f"Unique {column.replace('_', ' ')}:", unique_values)
```

```python
# Counting the number of missing values (NaN) in each column
data.isnull().sum()
```

We gonna create a bar chart that displays the percentage of null values in each column. This chart can help identify columns that have a large number of missing data points, allowing us to prioritize efforts to deal with the missing data :

First, we need to calculate the percentage of null values in each column:

```python
# Set the threshold percentage for null values
null_threshold = 1

# Calculate the percentage of null values for each column
null_percentage = data.isnull().sum() / len(data)  * 100

# Get the columns with null percentage greater than the threshold
null_cols = null_percentage[null_percentage > null_threshold]

# Print the null percentage for each column
for col in null_cols.index:
    # Format the output string to show column name and null percentage to two decimal places
    output_str = f'{col}: {null_percentage[col]:.2f}%'
    print(output_str)
```

Once we have calculated the percentage of null values for each column, we can create a bar chart to visualize this information :

```python
# Set the figure size
fig, ax = plt.subplots(figsize=(15, 9))

# Create a bar chart of the null percentage for each column
null_percentage.plot(kind='bar', ax=ax)
plt.title('Percentage of Null Values by Column')
plt.xlabel('Columns')
plt.ylabel('Percentage of Null Values')

plt.show()
```

**Data Visualization :**

We're going to visualize the inflation values of each year from 1970-2020 for every continent :

```python
# Extract the columns containing inflation data
inflation_columns = data_2.columns[5:57]

# Create a dataframe with the mean inflation for each continent
continent_mean_inflation = data.groupby('continent')[inflation_columns].mean()

# Create a line plot of the inflation ratios for each continent over the years
continent_mean_inflation.T.plot(kind='line')
plt.title('Inflation Ratio by Continent Over Time')
plt.xlabel('Years')
plt.ylabel('Inflation Ratio')
plt.show()
```

We're going to display the inflation values for each continent in their own separate plot, so it will be easier to observe and understand the data:

```python
# Extract the columns containing inflation data
inflation_columns = data.columns[5:58]

# Create a dataframe with the mean inflation for each continent (excluding South America)
continent_mean_inflation = data[data['continent'] != 'South America'].groupby('continent')[inflation_columns].mean()

# Create a sub-plot for each continent
continents = list(continent_mean_inflation.index)
fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(10, 12))

# Plot the inflation ratios for each continent over the years
for i, continent in enumerate(continents):
    row = i // 2
    col = i % 2
    continent_mean_inflation.loc[continent].plot(kind='line', ax=axes[row, col], legend=False)
    axes[row, col].set_title(continent)
    axes[row, col].set_xlabel('Years')
    axes[row, col].set_ylabel('Inflation Ratio')

plt.suptitle('Inflation Ratio by Continent Over Time (without South America)')
plt.tight_layout()
plt.show()
```

Finally, we are going to display the country with the highest inflation rate and the country with the lowest inflation rate for both 1970 and 2022:

Using the drop( ) method, we're going to drop all the columns of the dataset except the desired columns ('country', '1970', '2022') :

```python
data_2.drop(data_2.columns.difference(['country', '1970', '2022']), axis=1, inplace=True)
data_2
```

We are going to create a DataFrame that displays the countries with the highest inflation rates in 1970 and 2022, as well as the countries with the lowest inflation rates in 1970 and 2022 :

```python
# Find the country with the highest inflation in 1970
max_1970 = data_2.loc[data_2['1970'].idxmax()]

# Find the country with the highest inflation in 2022
max_2022 = data_2.loc[data_2['2022'].idxmax()]

# Create a DataFrame with the results
max_inflation_df = pd.DataFrame({
    'year': ['1970', '2022'],
    'country': [max_1970['country'], max_2022['country']],
    'inflation_rate': [max_1970['1970'], max_2022['2022']]
})

# Print the DataFrame
print(max_inflation_df)
```

```python
# Find the country with the lowest inflation in 1970
min_1970 = data_2.loc[data_2['1970'].idxmin()]

# Find the country with the lowest inflation in 2022
min_2022 = data_2.loc[data_2['2022'].idxmin()]

# Create a new DataFrame to store the results
min_inflation_df = pd.DataFrame({
    'year': [1970, 2022],
    'country': [min_1970['country'], min_2022['country']],
    'inflation_rate': [min_1970['1970'], min_2022['2022']]
})

# Print the results
print(min_inflation_df)
```
